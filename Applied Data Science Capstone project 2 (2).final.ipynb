###   Assign url to the wikipedia page

```python
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
```

```python
page = urllib.request.urlopen(url)
```

```python
soup = BeautifulSoup(page, "lxml")
```

### Use Prettify function of BeautifulSoup library to extract data

```python
print(soup.prettify())
```

```python
all_tables=soup.find_all("table")
all_tables
```

###  Extract table from wikipedia page

```python
right_table=soup.find('table', class_='wikitable sortable')
right_table
```

### Append table values into lists

```python
A=[]
B=[]
C=[]

for row in right_table.findAll('tr'):
    cells=row.findAll('td')
    if len(cells)==3:
        A.append(cells[0].find(text=True))
        B.append(cells[1].find(text=True))
        C.append(cells[2].find(text=True))
```

###   Add columns names to dataframe and Print dataframe

```python
df=pd.DataFrame(A,columns=['Postal Code'])
df['Borough']=B
df['Neighborhood']=C
df.head()
```

###  Filter only those values from dataframe , where Borough is assigned

```python
df_filtered=df[df["Borough"]!="Not assigned\n"]
df_filtered.head()
```

### Find if there are any duplicate Postal codes

```python
df_dup=df_filtered[df_filtered.duplicated(["Postal Code"])]
df_dup
```

###   Find if there is any neighborhood where value is 'Not assigned'

```python
df_f=df_filtered[df_filtered["Neighborhood"]=="Not assigned\n"]
df_f
```

###  Add new columns Latitude , Longitude to data

```python
df_new = df_new.reindex(df_new.columns.tolist() + ['Latitude','Longitude'], axis=1)
```

```python
df_new.head()
```

```python
geolocator = Nominatim(user_agent="can_explorer")
print(geolocator)
```

### Invoke Geolocator ,geocode function to get latitude, longitude

```python
index=0
df_new = df_new.replace('\n',' ', regex=True)
for index, row in df_new.iterrows():
        address = df_new.at[index,'Neighborhood']
        add1=address.rsplit(",")
        add2 =address + ' ,Canada'
        location = geolocator.geocode(add2)
        if (location):
            lat=location.latitude
            long=location.longitude
            df_new.at[index,'Latitude']= lat 
            df_new.at[index,'Longitude']= long
        else :
            address2= df_new.at[index,'Borough']
            loc=geolocator.geocode(address2)
            lat=loc.latitude
            long=loc.longitude
            df_new.at[index,'Latitude']= lat 
            df_new.at[index,'Longitude']= long
```

### Create Map of Canada - and mark markers

```python
geolocator = Nominatim(user_agent="can_explorer")
print(geolocator)
address = 'Ottawa, Canada'
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Canada are {}, {}.'.format(latitude, longitude))
```

```python
# create map of Canada using latitude and longitude values
map_canada = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(df_new['Latitude'], df_new['Longitude'], df_new['Borough'], df_new['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_canada)  
    
map_canada
```
